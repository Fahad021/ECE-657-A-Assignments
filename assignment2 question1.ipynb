1. Load a simple dataset and perform some basic data preprocessing to fill out ”unknowns”,
outliers or other invalid data. Explain what preprocessing was performed and why. Also,
change categorical data into numerical features using pandas.get dummies [5].

```python
import pandas as pd
df = pd.read_csv('bank-additional.csv', sep=';')
print(df.shape)
print(df.dtypes)
print(df.head())
print(df.groupby('y').size())
```

```python
df.isnull().sum()
```

```python
obj_df = df.select_dtypes(include=['object']).copy()
obj_df.head()
```

```python
num_df1 = df.select_dtypes(include=['int64']).copy()
num_df1.head()
```

```python
num_df1.shape
```

```python
num_df1 = num_df1.apply(pd.to_numeric)
num_df1 = num_df1.astype('float64')
num_df1.dtypes
```

```python
num_df2 = df.select_dtypes(include=['float64']).copy()
num_df2.head()
```

```python
num_df2.dtypes
```

```python
num_total = pd.concat([num_df1,num_df2], axis=1, sort=False)
num_total.head()
```

```python
num_total.dtypes
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy 

array = num_total.values

scaler = MinMaxScaler(feature_range=(0, 1))
rescaled = scaler.fit_transform(array)
# summarize transformed data
#set_printoptions(precision=3)
print(rescaled[0:5,:])
```

```python
columns= list(num_total)
print(columns)
```

```python
num_total = pd.DataFrame(data=rescaled, columns=columns)
num_total.describe()
```

```python
num_total.head(5)
```

```python
#https://datascience.stackexchange.com/questions/14069/mass-convert-categorical-columns-in-pandas-not-one-hot-encoding
char_cols = obj_df.dtypes.pipe(lambda x: x[x == 'object']).index
for c in char_cols:
    obj_df[c] = pd.factorize(obj_df[c])[0]
```

```python
obj_df.head()
```

```python
df1 = pd.concat([num_total,obj_df], axis=1, sort=False)
df1.head()
```

```python
df1.describe()
```

3. Apply classification using Decision Trees (DT), Random Forests (RF) and Neural Networks (NN) and run using standard libraries in your language or choice. Indicate the classification properties (example: depth of tree, size of neural network, ensembles for RF) you have chosen and justify. Briefly describe the algorithms employed by the libraries you are using (example ID3, C4.5, C5.0 and CART for DT). Make sure to run the classification on 10 features or more [10]

```python
from matplotlib import pyplot
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

dataframe = df1
array = dataframe.values
X = array[:,0:20]
Y = array[:,20]
```

```python
X
```

```python
Y
```

```python
num_folds = 10
seed = 7
kfold = KFold(n_splits=num_folds, random_state=seed)
```

```python
#1
model = DecisionTreeClassifier()
results = cross_val_score(model, X, Y, cv=kfold)
print("Accuracy: %.3f%% (%.3f%%)" % (results.mean()*100.0, results.std()*100.0))
```

```python
#2
num_trees = 100
max_features = 3
kfold = KFold(n_splits=10, random_state=7)
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model, X, Y, cv=kfold)
print("Accuracy: %.3f%% (%.3f%%)" % (results.mean()*100.0, results.std()*100.0))
```
